1. Write a "counter" class that can be incremented up to a specified maximum value, will print an error if an attempt is made to increment beyond that value, and allows reseting the counter. 

```python
class Counter:
    def __init__(self,max_value):
        self.max_value=max_value
        self.value=0

    def increment(self):
        if self.value < max_value:
            self.value+=1
        else:
            print("Error!! Max value reached")

    def reset(self):
        self.value=0

    def return_value(self):
        return self.value
max_value = 7
counter_instance = Counter(max_value)

for _ in range(max_value + 1):
    counter_instance.increment()
    print("Current Value:", counter_instance.return_value())

counter_instance.reset()
print("After Reset, Current Value:", counter_instance.return_value())
```

```python


class RasterDrawing:
    def __init__(self, width, height):
        self.canvas = Canvas(width, height)
        self.shapes = []

    def add_shape(self, shape):
        self.shapes.append(shape)

    def paint_drawing(self):
        # Clear the canvas
        self.canvas = Canvas(self.canvas.width, self.canvas.height)


        # Display the canvas
        print("Drawing:")
        self.canvas.display()

    def modify_drawing(self):
        # Modify the drawing (for demonstration purposes)
        if len(self.shapes) > 0:
            # Move the first shape to a new position
            first_shape = self.shapes[0]
            first_shape.move_to(new_x=15, new_y=8)

    def paint_modified_drawing(self):
        # Paint the modified drawing
        self.paint_drawing()

    def save_to_file(self, filename):
        with open(filename, "w") as file:
            file.write(repr(self.shapes))

    def load_from_file(self, filename):
        if os.path.exists(filename):
            with open(filename, "r") as file:
                shapes_repr = file.read()
                self.shapes = eval(shapes_repr)

# Example usage
if __name__ == "__main__":
    # Create a RasterDrawing
    drawing = RasterDrawing(width=30, height=15)

    # Add shapes to the drawing
    rectangle = Rectangle(length=8, width=4, x=5, y=5)
    circle = Circle(radius=6, x=18, y=10)

    drawing.add_shape(rectangle)
    drawing.add_shape(circle)

    # Paint the initial drawing
    drawing.paint_drawing()

    # Save the drawing to a file
    drawing.save_to_file("drawing.txt")

    # Modify the drawing
    drawing.modify_drawing()

    # Paint the modified drawing
    drawing.paint_modified_drawing()

    # Load the drawing from the file
    loaded_drawing = RasterDrawing(width=30, height=15)
    loaded_drawing.load_from_file("drawing.txt")

    # Paint the loaded drawing
    loaded_drawing.paint_drawing()
```
